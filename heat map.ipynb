```python
import numpy as np
import matplotlib.pyplot as plt
X_pos = 0
Y_pos = 0
X_direct = 1
Y_direct = 1
fig, ax = plt.subplots(figsize=(12, 7))
ax.quiver(X_pos, Y_pos, X_direct, Y_direct)
ax.set_title('Quiver plot with one arrow')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X_pos = [0, 2]
Y_pos = [0, 1]
X_direct = [1, -1]
Y_direct = [1, -1]

fig, ax = plt.subplots(figsize=(12, 7))
ax.quiver(X_pos, Y_pos, X_direct, Y_direct)
ax.set_title('Quiver plot with two arrows in different directions')
plt.show()
```
