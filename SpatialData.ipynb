```python
import folium
mymap = folium.Map(location = [13.172907827355605, 77.5370178073665],
                   width = 950,
                   height = 550,
                   zoom_start = 12,
                   tiles = 'openstreetmap')
mymap
```

```python

marker = folium.Marker(location=[13.245438283707708, 77.70658372072477],
                       popup='Shahe darbar :devanahalli',
                       icon=folium.Icon(color='red')
)


mymap.add_child(marker)
```
