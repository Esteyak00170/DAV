```python
import folium
mymap = folium.Map(location = [13.172907827355605, 77.5370178073665],
                   width = 950,
                   height = 550,
                   zoom_start = 12,
                   tiles = 'openstreetmap')
mymap
```

```python

marker = folium.Marker(location=[18.956346135545505, 72.83255470108985],
                       popup='Delhi Zaika Bhendi Bazaar',
                       icon=folium.Icon(color='red')
)


mymap.add_child(marker)
```

```python
import folium

mymap = folium.Map(location=[13.013371307375166, 77.58810505900468], zoom_start=12)

image_url = 'https://en.wikipedia.org/wiki/Masjid_al-Haram#/media/File:Great_Mosque_of_Mecca1.jpg'

icon = folium.CustomIcon(
    image_url,
    icon_size=(50, 50),
    icon_anchor=(25, 25),
)

marker = folium.Marker(
    location=[21.440166759972765, 39.81351969980192],
    popup='Makkha',
    icon=icon
)

mymap.add_child(marker)

mymap
```
