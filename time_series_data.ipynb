```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2023-01-01","2023-02-01","2023-03-01"]),
    "value":[10,15,22,28,35,42,48,55,60,68,75,82]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o')
plt.xlabel("date")
plt.ylabel("value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```
