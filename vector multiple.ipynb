```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.arange(-2, 2, 0.2),
                   np.arange(-2, 2, 0.2))
u = np.sin(np.pi * x) * np.cos(np.pi * y)
v = -np.cos(np.pi * x) * np.sin(np.pi * y)

plt.figure()
plt.quiver(x, y, u, v)
plt.title('Quiver Plot with Multiple Arrows Forming Patterns')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
u = np.sin(X**2 + Y**2)
v = np.cos(X**2 - Y**2)

plt.figure()
plt.quiver(X, Y, u, v)
plt.title('Quiver Plot with Arrows')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
u = np.sin(X) * np.cos(Y)
v = -np.cos(X) * np.sin(Y)

plt.figure()
plt.quiver(X, Y, u, v, scale=20,cmap='viridis')
plt.title('Quiver Plot with Scaling')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
u = np.sin(X) * np.cos(Y)
v = -np.cos(X) * np.sin(Y)

speed = np.sqrt(u**2 + v**2)
color = np.arctan2(v, u)

plt.figure()
plt.quiver(X, Y, u, v, speed, cmap='viridis')
plt.title('Quiver Plot with Colors')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.colorbar(label='Speed')
plt.show()
```
